```python
# -*- coding: utf-8 -*-
"""
Created on Sat Sep 28 20:00:56 2024
영어단어, 한글 뜻으로 구분된
엑셀 파일 읽기
한개 파일로 쭉 읽기 파일 한개인 이유. 나중에 한번에 보기 편하라고
최근 회차 단어 + 이전회차 랜덤선택 (총 100개)
단어 순서 랜덤으로 배열하기
@author: young
"""
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime


file_path = 'wordsheet.xlsx'  # Update with your file path
df = pd.read_excel(file_path, engine='openpyxl')
    
df 
```

```python
회차=2
total_pick_num=100
```

```python
cur_count = df[df['chapter'] == 회차].shape[0]

review_count = df[df['chapter'] <= 회차].shape[0] #복습 가능한 단어 개수
if(review_count<total_pick_num): #뽑아야 하는 복습 단어 수가 뽑을수 있는 복습단어 수보다 작으면 뽑아야 하는 개수 축소
    total_pick_num=df.shape[0]
review_word_num=total_pick_num-cur_count #복습 단어 뽑는 개수

per_chapter=review_word_num//(회차-1) #회차별 뽑아야 하는 단어 개수

chapters_to_include_all = [회차]
def select_rows(group):
    if group['chapter'].iloc[0] in chapters_to_include_all:
        return group  # 선택된 챕터의 모든 행을 포함
    else:
        return group.sample(min(len(group), per_chapter))  # 나머지 챕터는 랜덤 샘플링

# 그룹화하여 각 챕터 처리
selected_rows = df.groupby('chapter').apply(select_rows).reset_index(drop=True)

shuffled_df = selected_rows.sample(frac=1).reset_index(drop=True)

shuffled_df
```

```python
shuffled_df.drop(['chapter', 'korean'], axis=1, inplace=True)
```

```python
import matplotlib.font_manager as fm
# Figure 및 Axis 생성
fig, ax = plt.subplots(figsize=(8, 4))

font_path = 'C:\\Windows\\Fonts\\batang.ttc'  # Windows 예시

# 폰트 프로퍼티 생성
font_prop = fm.FontProperties(fname=font_path)
# 테이블 추가
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=shuffled_df.values, colLabels=shuffled_df.columns, cellLoc='right', loc='left')

now = datetime.now()

# 원하는 형식으로 출력 (예: YYYY-MM-DD HH:MM:SS)
formatted_now = now.strftime("%Yy%mm%dd %H %M %S")

# 테이블 스타일 조정
table.scale(1, 1.5)
for (i, j), cell in table.get_celld().items():
    # 각 셀의 텍스트 속성을 변경하여 폰트를 설정
    if cell.get_text() is not None:  # 텍스트가 있는 경우
        cell.set_text_props(fontproperties=font_prop)
# 이미지로 저장
plt.title(str(회차)+"회차", fontproperties=font_prop)
plt.savefig(f"{회차}회차 {formatted_now}.png", bbox_inches='tight', dpi=300)
plt.close()
```
